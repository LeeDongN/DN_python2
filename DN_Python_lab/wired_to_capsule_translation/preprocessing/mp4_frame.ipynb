```python
data_vidcap = cv2.VideoCapture(r'C:\DATA_Python\cap_endoscope.mp4')

count = 0

while(data_vidcap.isOpened()):
    ret, image = data_vidcap.read()
    
    cv2.imwrite(r'C:\DATA_Python\endoscope\cap_endoscope_frame\%d.jpg' % count, image)
    
    print('Saved frame%d.jpg' % count)
    count += 1

data_vidcap.release()
```

```python
def video2frame(invideofilename, save_path):
    vidcap = cv2.VideoCapture(invideofilename)
    count = 0
    while True:
        success,image = vidcap.read()
        if not success:
            break
            
        print ('Read a new frame: ', success)
        fname = "{}.jpg".format("{0:05d}".format(count))
        cv2.imwrite(save_path + fname, image) # save frame as JPEG file
        count += 1
    print("{} images are extracted in {}.". format(count, save_path))
```

```python
video2frame(r'C:\DATA_Python\endoscope\cap_endoscope.mp4',r'C:\DATA_Python\endoscope\cap_endoscope_frame')
```
