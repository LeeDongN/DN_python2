```python

img_color = cv2.imread('C:\\program1\\image_raw\\01-0004.png')
img_gray = cv2.imread('C:\\program1\\image_raw\\01-0004.png', cv2.IMREAD_GRAYSCALE)
        
#이미지의 윤곽선을 딴다
blur = cv2.GaussianBlur(img_gray, ksize = (5, 5), sigmaX = 0)
ret, thresh1 = cv2.threshold(blur, 127, 255, cv2.THRESH_BINARY)

edged = cv2.Canny(blur, 10, 250)


contours, _ = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours_image = cv2.drawContours(img_color, contours, -1, (0,255,0), 3)
cv2.imshow('contours_image', contours_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
